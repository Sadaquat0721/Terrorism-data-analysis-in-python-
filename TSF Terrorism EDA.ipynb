```python
pd.set_option('display.max_columns',500)
```

```python
df["gname"].value_counts()[:20]
```

```python
attacking_group=df['gname'].value_counts()[1:11].reset_index()
attacking_group.columns=['attacking_grp_name','Total_attacks']
sns.barplot(x=attacking_group.attacking_grp_name,y=attacking_group.Total_attacks,palette='cubehelix_r',edgecolor=sns.color_palette('YlOrRd',10))
plt.xticks(rotation=60)
plt.title('Total attacks by different groups ')
plt.show()
```

```python
country_wise_attacks=df['country_txt'].value_counts()[:10].reset_index()
country_wise_attacks.columns=['Country','Total_attacks']
sns.barplot(x=country_wise_attacks.Country,y=country_wise_attacks.Total_attacks,palette='RdYlBu')
plt.xticks(rotation=60)
plt.title('Country wise total attacks')
plt.show()
```

```python
region_wise_attacks=df['region_txt'].value_counts().reset_index()
region_wise_attacks
```

```python
region_wise_attacks.columns=['Region','Total_attacks']
sns.barplot(x=region_wise_attacks.Region,y=region_wise_attacks.Total_attacks,palette='BuGn_r',edgecolor=sns.color_palette('dark',10))
plt.xticks(rotation=60)
plt.ylabel('Total Attacks ')
plt.title('Total Attacks Region wise')
plt.show()
```

```python
city_attacked_most=df['city'].value_counts()[:15].reset_index()
city_attacked_most
```

```python
city_attacked_most=city_attacked_most[1:16]
city_attacked_most.columns=['City','Total_attacks']
sns.barplot(x=city_attacked_most.City,y=city_attacked_most.Total_attacks,palette='PiYG_r',edgecolor=sns.color_palette('magma',20))
plt.xticks(rotation=30)
plt.title('City Targeted Most by Terror groups')
plt.show()
```

```python
df['Casualities']=df['nkill']+df['nwound']
```

```python
yearly_casualities=df.groupby('iyear').Casualities.sum().reset_index()
yearly_casualities.columns=['Year','Casualities']
sns.barplot(x=yearly_casualities.Year,y=yearly_casualities.Casualities,palette='YlOrBr')
plt.xticks(rotation=90)
plt.ylabel('Yearly Casualities')
plt.show()
```

```python
plt.subplots(figsize=(20,8))
sns.countplot(x=df.iyear,data=df,palette='RdYlGn_r')
plt.xticks(rotation=90)
plt.title('Terrorist Activities Each Year')
plt.show()
```
